```python
# Question 1 & 2: Create the table and load the CSV data natively using Apache Hive
!$HIVE_HOME/bin/hive -e " \
CREATE TABLE IF NOT EXISTS employee ( \
    emp_id INT, \
    name STRING, \
    age INT, \
    department STRING, \
    salary DOUBLE \
) \
ROW FORMAT DELIMITED \
FIELDS TERMINATED BY ',' \
STORED AS TEXTFILE \
TBLPROPERTIES ('skip.header.line.count'='1'); \
\
LOAD DATA LOCAL INPATH '/content/employee_data.csv' INTO TABLE employee; \
\
SELECT * FROM employee LIMIT 3; \
"
```

```python
# Question 3: Find the average salary of employees natively in Hive
!$HIVE_HOME/bin/hive -e "SELECT AVG(salary) AS average_salary FROM employee;"
```

```python
# Question 4: Extract records where age is greater than 30
!$HIVE_HOME/bin/hive -e "SELECT emp_id, name, age, department, salary FROM employee WHERE age > 30;"
```

```python
# Question 5: Create department table, insert reference records, and execute the INNER JOIN
!$HIVE_HOME/bin/hive -e " \
CREATE TABLE IF NOT EXISTS department ( \
    department STRING, \
    department_name STRING \
) \
ROW FORMAT DELIMITED \
FIELDS TERMINATED BY ','; \
\
INSERT INTO TABLE vdepartment VALUES \
('HR', 'Human Resources'), \
('Finance', 'Finance and Accounts'), \
('IT', 'Information Technology'); \
\
SELECT \
    e.emp_id, \
    e.name, \
    e.department, \
    d.department_name, \
    e.salary \
FROM \
    employee e \
INNER JOIN \
    department d \
ON \
    e.department = d.department; \
"
```
