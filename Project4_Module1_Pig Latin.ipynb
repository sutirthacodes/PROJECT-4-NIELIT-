```python
#Question 1: Write a Pig script to load the employee_data.csv file and filterthe rows where the salary is greater than 60000.


employee = LOAD 'employee_data.csv'
USING PigStorage(',')
AS(id:int,name:chararray,department:chararray,salary:int);

result = FILTER employee BY salary > 60000;

DUMP result;
```

```python
!pig-0.17.0/bin/pig -x local filter.pig
```

```python
#Question 2: Use Pig to group the employee_data.csv by department andcalculate the sum of salary.


employee = LOAD 'employee_data.csv'
USING PigStorage(',')
AS(id:int,name:chararray,department:chararray,salary:int);

group_data = GROUP employee BY department;

result = FOREACH group_data
GENERATE group, SUM(employee.salary);

DUMP result;
```

```python
!pig-0.17.0/bin/pig -x local group.pig
```

```python
# Question 3: Write a Pig Latin command to join the employee_data.csv anddepartment_data.csv on the department column. 


employee = LOAD 'employee_data.csv'
USING PigStorage(',')
AS(id:int,name:chararray,department:chararray,salary:int);


dept = LOAD 'department_data.csv'
USING PigStorage(',')
AS(department:chararray,manager:chararray);


joined = JOIN employee BY department,
dept BY department;


DUMP joined;
```

```python
!pig-0.17.0/bin/pig -x local join.pig
```

```python
# Question 4: How do you remove duplicate rows from a dataset using Pig?

employee = LOAD 'employee_data.csv'
USING PigStorage(',');

unique_data = DISTINCT employee;

DUMP unique_data;
```

```python
!pig-0.17.0/bin/pig -x local duplicate.pig
```

```python
# Question 5: Write a Pig script to count the number of employees in each department. 

employee = LOAD 'employee_data.csv'
USING PigStorage(',')
AS(id:int,name:chararray,department:chararray,salary:int);


group_department = GROUP employee BY department;


result = FOREACH group_department
GENERATE group, COUNT(employee);


DUMP result;
```

```python
!pig-0.17.0/bin/pig -x local count.pig
```
